```python
import json
```

```python
f = open('words.txt','w')
data = {}
json.dump(data,f)
f.close()

while True:
    x = int(input(''' Welcome to the English Dictionary app !
    Select the below given options :
    1=Add a new word
    2=Find the meaning 
    3=Update a word
    4=Exit
    '''))
    
    if x == 1:
        f = open('words.txt','w')
        w=input('Enter a word')
        m=input('Enter a meaning -')
        data[w]=m
        json.dump(data,f)
        f.close()
    
    elif x == 2:
        with open('words.txt','r+') as f:
            data = json.load(f)
            print(data)
            word = input('Enter a word -')
            if word in data:
                print(data[word])
            else:
                print('Word does not exist')
            f.close()
            
    elif x == 3:
        with open('words.txt','w') as f:
            print(data)
            word1 = input('Enter a word -')
            if word1 in data.keys():
                m1 = input('Enter the meaning -')
                data[word1] = m1
                json.dump(data,f) 
                f.close()
                
    elif x == 4:
        print('Thank you for using the dictionary')
        break
```
